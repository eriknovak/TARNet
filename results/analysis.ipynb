```python
import pandas as pd
```

```python
acc_df = pd.read_csv("./sparsity=0.8_batch/accuracy_lr=0.001.csv", names=["dataset", "model", "k_fold", "accuracy"])
```

```python
stats_df = acc_df.groupby("dataset") \
    .agg(mean=("accuracy", "mean"), std=("accuracy", "std"))
```

```python
stats_df.sort_values(by="dataset", ascending=True)
```
